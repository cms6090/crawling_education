```python
import re
for i in range(len(docs)):
    docs['doc'][i] = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣]", "", str(docs['doc'][i]))
docs
```

```python
len(docs)
```

```python
for i in range(len(docs)):
    if (len(docs['doc'][i]) < 2 or docs['doc'][i].isspace() == True): #문서 내용이 두글자 미만이나 공백으로만 되있는 경우 결측값 처리
        docs = docs.drop(i)   #(삭제할 인덱스, axis)  #axis = 1일때 열 삭제, 0일때 행 삭제
print("ok")
```

```python
docs = docs.reset_index(drop=True)
docs
```

```python
from tqdm import tqdm
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
token_list = []
token_noun = []
for i in tqdm(range(len(docs))):
    pos = komoran.pos(u'{}'.format(docs['doc'][i]))
    noun = list(term for term in komoran.nouns(u'{}'.format(docs['doc'][i])) if len(term) >1)
    token_list.append(pos)
    token_noun.append(noun)
```

```python
token_noun
```

```python
f = open(file_path + "stopwords-ko.txt", "r", encoding = "UTF-8")
st = f.readlines()
f.close()
st
```

```python
stw = []
for i in range(0, len(st)):
    stw.append(st[i].rstrip('\n'))
stw
```

```python
for word in stw:
    for i in range(0, len(token_noun)):
        while word in token_noun[i]:
            token_noun[i].remove(word)
```

```python
docs['token_list_pos'] = token_list
docs['token_noun'] = token_noun
```

```python
import pickle
f = open(file_path + "total_doc.pkl", "wb")
pickle.dump(docs, f)
f.close()
```

## TF-IDF

```python
doc_noun = []
for i in range(0,len(data['token_noun'])):
    doc_noun.append(' '.join(data['token_noun'][i]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer(min_df = 0.01, max_df = 0.95)
X = vec.fit_transform(doc_noun)
```

```python
print(X)
```

## K-MEANS

```python
from sklearn.cluster import KMeans
def elbow(X):
    sse = []
    
    for i in range(1,10):
        km = KMeans(n_clusers=i, algorithm='auto', random_state = 0)
        km.fit(X)
        sse.append(km.inertia_)
        print(i)
        
    plt.plot(range(1,10), sse, marker='o')
    plt.xlabel('K')
    plt.ylabel('SSE')
    plt.xticks(range(1,10))
    plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, algorithm = 'auto', random_state=0)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vec.get_feature_names()
for i in range(3):
    print("Cluster %d: " %i)
    for ind in order_centroids[i, :30]:
        print(' %s' %terms[ind])
    print('\n')
```

```python
from gensim import corpora, models
noun_dic = corpora.Dictionary(data['token_noun'])
noun_dic.token2id
```

```python
import numpy
from gensim import corpora, models
```

```python
corpus = [noun_dic.doc2bow(text) for text in data['token_noun']]
corpus
```

```python
import gensim
from gensim.models import CoherenceModel

Lda = gensim.models.ldamodel.LdaModel
perplexity_score = []
coherence_score = []

for i in range(1,10):
    ldamodel=Lda(corpus, num_topics=i, id2word=noun_dic, passes=15, random_state=0)
    perplexity_score.append(ldamodel.log_perplexity(corpus))
    coherence_score.append(CoherenceModel(model=ldamodel, texts=data['token_noun'], dictionary=noun_dic, coherence='c_v').get_coherence())
    
    print(i, 'process complete')
```

```python
plt.plot(range(1,10),perplexity_score,'r',marker='^')
plt.xlabel("number of topics")
plt.ylabel("perplexity score")
plt.show()
```

```python
plt.plot(range(1,10), coherence_score,'b',marker='o')
plt.xlabel("number of topics")
plt.ylabel("coherence score")
plt.show()
```

```python
noun_lda = Lda(corpus, num_topics=2, id2word=noun_dic, passes=15, random_state=0)
topics = noun_lda.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

```python
import pyLDAvis
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook()
lda_viz = pyLDAvis.gensim_models.prepare(noun_lda, corpus, noun_dic)

pyLDAvis.display(lda_viz)
```

```python
noun_dtm = [noun_dic.doc2bow(text) for text in data['token_noun']]
ldamodel = Lda(noun_dtm, num_topics = 3, id2word = noun_dic, passes = 10, random_state = 0)

topics = []
for i in range(len(noun_dtm)):
    prop_sort = []
    topic_sort = []
    for topic, prop in ldamodel.get_document_topics(noun_dtm)[i]:
        prop_sort.append(prop)
        topic_sort.append(topic)
    topics.append(topic_sort[prop_sort.index(max(prop_sort))])
    
data['topic'] = topics
```
